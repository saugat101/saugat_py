```python
def cal(l,b,h):
    a=l*b
    v=h*a
    return a,v
area,volume=cal(10,10,10)
print("Area = ",area)
print("Volume = ",volume)
```

```python
print("Calculator")
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
a=float(input("Enter the first of number two calculate : "))
b=float(input("Enter the second of number two calculate : "))
o=input("Enter the operator : ")
if o=="+":
    print("The sum =",add(a,b))
elif o=="-":
    print("The difference =",sub(a,b))
elif o=="*":
    print("The product =",mul(a,b))
elif o=="/":
    print("The quotiont =",div(a,b))
elif o=="%":
    print("The remainder =",a%b)
else:
    print("Enter valid operator")
```
